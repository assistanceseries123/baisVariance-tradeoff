```python
import numpy as np
import matplotlib.pyplot as plt

#set the parameters
num_samples=10000
sample_size=300
distribution_range=(0,1)

#Generate samples for the uniform distribution
samples=np.random.uniform(distribution_range[0],distribution_range[1],(num_samples,sample_size))
#calculate the sample means
sample_means=np.mean(samples,axis=1)

#plot the histogram of the sample means
plt.hist(sample_means,bins=20,density=True,edgecolor='black')
plt.title("HIstogram of the sample means")
plt.xlabel("Sample mean")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#set the parameters
num_samples=1000
sample_size=50
lambda_param=2
#Generate sample from an exponential distribution
samples=np.random.exponential(scale=1/lambda_param,size=(num_samples,sample_size))
#calculate the sample means
sample_means=np.mean(samples,axis=1)

#plot the histogram of the sample means
plt.hist(sample_means,bins=30,density=True,edgecolor='black')
plt.title("Histogram of sample means of the exponential distribution")
plt.xlabel("Sample mean")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#set the parameter
num_samples=1000
sample_size=50
#poisson distribution parameters
poisson_lambda=5
#gamma distribution parameters
gama_shape=2
gamma_scale=1
#Binomial distribution parameters
binomial_n=10
binomial_p=0.5
#generate samples for the distribution
poisson_sample=np.random.poisson(lam=poisson_lambda,size=(num_samples,sample_size))
gamma_sample=np.random.gamma(shape=gama_shape,scale=gamma_scale,size=(num_samples,sample_size))
binomial_samples=np.random.binomial(n=binomial_n,p=binomial_p,size=(num_samples,sample_size))
#calculate the sample means
poisson_means=np.mean(poisson_sample,axis=1)
gamma_means=np.mean(gamma_sample,axis=1)
binomial_means=np.mean(binomial_samples,axis=1)
#plot the histogram of the sample means
fig,axs=plt.subplots(3,1,figsize=(8,12))
#poisson distribution
axs[0].hist(poisson_means,bins=30,density=True,edgecolor='black')
axs[0].set_title("Histogram of sample means (poissons distribution)")
axs[0].set_xlabel("sample mean")
axs[0].set_ylabel('Density')
#gamma distribution
axs[1].hist(gamma_means,bins=30,density=True,edgecolor='black')
axs[1].set_title("Histogram of sample means (gamma distribution)")
axs[1].set_xlabel("sample mean")
axs[1].set_ylabel("Density")

#Binomial distribution
axs[2].hist(binomial_means,bins=30,density=True,edgecolor='black')
axs[2].set_title("Histogram of sample means (binomial distribution)")
axs[2].set_xlabel("Sample mean")
axs[2].set_ylabel("Density")
```

```python
import numpy as np
#set the parameter
num_samples=10000
sample_size=50

#Gamma distribution parameters
gamma_shape=2
gamma_scale=1
#Calculate the theoretical mean and variance
theoretical_mean=gama_shape*gamma_scale
theoretical_variance=(gamma_shape*gamma_scale**2)
#Generate sample for the gamma distribution
samples=np.random.gamma(shape=gamma_shape,scale=gamma_scale,size=(num_samples,sample_size))
#Calculate the sample means
sample_means=np.mean(samples,axis=1)
#Calculate the emprical mean and variance of the sample means
emprical_mean=np.mean(sample_means)
emprical_variance=np.var(sample_means)
#compare the theoretical and the emprical values
print(f"Theoretical mean:{theoretical_mean:.4f}")
print(f"Emprical mean:{emprical_mean:.4f}")
print(f"\n")
print(f"Theoretical variance:{theoretical_variance:.4f}")
print(f"Emprical variance:{emprical_variance:.4f}")
```

```python
import numpy as np

#set the parameter
num_samples=10000
sample_size=50
#log normal distribution parameter
mu=0.5
sigma=0.7
#Calculate the theoretical mean and variance
theoretical_mean=np.exp(mu+(sigma**2)/2)
theoretical_variance=(np.exp(sigma**2)-1)*np.exp(2*mu+sigma**2)
#Genereate the sample for the log normal distribution
samples=np.random.lognormal(mean=mu,sigma=sigma,size=(num_samples,sample_size))
#Calculate the sample means
sample_means=np.mean(samples,axis=1)
#Calculate the emprical means and variance of the sample means
emprical_mean=np.mean(sample_means)
emprical_variance=np.var(sample_means)
#Compare the theoretical and emprical values
print(f"Theoretical mean:{theoretical_mean:.4f}")
print(f"Emprical mean:{emprical_mean:.4f}")
print(f"\n")
print(f"Theoretical variance:{theoretical_variance:.4f}")
print(f"Emprical variance:{emprical_variance:.4f}")
```

```python
#Central limit theorem case study
import numpy as np
#Set the parameter
population_size=100000
sample_size=50
num_samples=100
#Generate the random representative sample of salaries (in thousand)
#You should replace with actual colleted salary data
np.random.seed(42) #Setting the seed for the reproducibility
population_salaries=np.random.lognormal(mean=4.5,sigma=0.8,size=population_size)
#Generate multiple samples and calculate the sample means and standard deviation
sample_means=[]
sample_std_devs=[]
for _ in range(num_samples):
    sample_salaries=np.random.choice(population_salaries,size=sample_size)
    sample_means.append(np.mean(sample_salaries))
    sample_std_devs.append(np.var(sample_salaries))
#calculate the average of the sample means and the standard error
average_sample_means=np.mean(sample_means)
standard_error=np.std(sample_means)/np.sqrt(num_samples)
#Calculate the 95% confidence interval
margin_of_error=1.96*standard_error
lower_limit=average_sample_means-margin_of_error
upper_limit=average_sample_means+margin_of_error
#Report the result
print(f"Estimated average salary (in thousand):{average_sample_means:.2f}")
print(f"95% confidence interval (in thousand):{lower_limit:.2f},{upper_limit:.2f}")
```
